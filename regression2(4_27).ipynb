3차원 plot 그래프 (책 p 186 참조)

```python

from mpl_toolkits.mplot3d import Axes3D 

def show_data2(ax, x0,x1, t):
  for i in range(len(x0)):
    ax.plot([x0[i],x0[i]],[x1[i],x1[i]],[120,t[i]], color='gray')  #실선
    ax.plot(x0, x1, t, 'ob')  #점

plt.figure()
ax=plt.subplot(1,1,1,projection='3d')    
show_data2(ax, X0,X1,T)
plt.show()
```

세 변수의 관계는 ?
  모델 함수 y=w_0 x+ w_1x_1+w_2        (예 w_0=1.5, w_1=1, w_2=90)  책 p189





```python
def show_plane(ax, w):
  px0=np.linspace(5,25,5)
  px1=np.linspace(35,75,5)
  px0, px1=np.meshgrid(px0,px1)
  z=w[0]*px0+w[1]*px1+w[2]
  ax.plot_surface(px0, px1, z, alpha=0.3, color="blue")

plt.figure(figsize=(6,5))
ax=plt.subplot(1,1,1,projection='3d')   
W=[1.5, 1, 90] 
show_plane(ax, W)
show_data2(ax, X0, X1,T)
plt.show()
```

행렬을 이용한 연립방정식  (X^T X) W= X^T T  을 이용하여 아래 데이터의 회귀모델의 매개변수 벡터 W 구하기 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

X0=np.array([15,23,5,9,7,10,14]) #나이
X1=np.array([70,58,37,57,57,41,58]) #몸무게
T=np.array([171,161,129,160,155,141,154]) # 키
```

```python
X=np.zeros((7,3))
X[:,0]=X0
X[:,1]=X1
X[:,2]=1
print(X)
```

```python
A = np.linalg.inv(np.dot(X.T,X))
B = np.dot(X.T,T)
W = np.dot(A,B)
print(W)
```

회귀모델의 매개변수를 구했다면 위의 3차원 figure 코드를 참고하여 데이터와 회귀모델을 그리시오.

```python

plt.figure(figsize=(10,8))
ax=plt.subplot(1,1,1,projection='3d')
show_plane(ax, W)
show_data2(ax, X0, X1,T)
plt.show()
```
